# Abstract

Childcare affordability is a pressing concern across the United States, with high and uneven costs significantly influencing family decision-making, workforce participation, and access to early education. These substantial costs often dictate where families choose to live, how much parents are able to work, and the type of care they can afford. Despite the importance of cost transparency, many families lack access to reliable, localized data on childcare prices, making it difficult to compare options, identify affordable alternatives, or plan effectively.

To address this gap, this project introduces SmartCare Locator, a predictive and personalized web-based application designed to help families explore and compare childcare affordability across U.S. counties. The tool estimates county-level childcare prices and recommends more affordable regions based on household income, child age group, and preferred provider type.

Leveraging the comprehensive National Database of Childcare Prices (NDCP), which integrates detailed pricing data from state market rate surveys with crucial county-level demographic indicators derived from the American Community Survey (ACS), the SmartCare system applies machine learning techniques to generate price estimates and affordability rankings. The resulting application offers a practical, data-driven solution that enables families to make more informed, budget-conscious childcare decisions with greater confidence and clarity.

# Introduction

The high cost of childcare in the United States continues to impose a substantial burden on families, directly affecting economic stability, career decisions, and access to early education opportunities. Across regions, childcare prices vary widely and unpredictably, with some families paying more for care than for housing or higher education. These cost disparities can determine where families live, whether both parents can participate in the workforce, and what level of care their children receive. Despite the central role of childcare in shaping these life decisions, many households lack access to reliable, localized data that could help them evaluate affordability and plan accordingly.

While several states conduct market rate surveys (MRS) to assess childcare pricing, these surveys are not uniform in frequency, geographic coverage, or methodology. As a result, data availability is often inconsistent, particularly at the county level, where families are most affected by cost differences. This lack of transparency limits not only public understanding of childcare markets but also the ability of families to make informed, cost-effective choices. In many cases, there is no simple way to determine:

* Whether childcare would be more affordable in a neighboring county
* How cost varies between provider types (center-based vs. home-based care)
* What affordability looks like in relation to a household’s income and child age group

To address these gaps, this project introduces SmartCare Locator, a predictive and personalized web application developed to assist families in navigating childcare affordability. The system is designed to perform two core functions:

* **Price Prediction:** Estimate weekly childcare costs for counties with missing or outdated data using machine learning models trained on demographic and economic indicators
* **Affordability Recommendation:** Rank and recommend counties based on a personalized affordability score that factors in household income, child age group, and care type preference

By combining these two components, the tool simplifies complex childcare pricing data into user-friendly, actionable insights. It eliminates the need for users to interpret raw statistics or cross-reference incomplete datasets, and instead delivers targeted recommendations based on their individual profiles.

The expected outcomes of the project include:

* A fully functional, publicly accessible web application built using Streamlit
* Accurate, data-driven childcare price estimates for underreported counties
* Ranked lists of affordable counties tailored to household financial situations
* A clear, interpretable interface designed specifically for families—not policymakers or researchers

Through its predictive and recommendation capabilities, SmartCare Locator contributes a practical, family-centered solution that supports more informed and confident decision-making around one of the most financially impactful aspects of raising children in the United States.

# Project Methodology

The development of the SmartCare Locator involves four major phases: data preprocessing, predictive modeling, affordability scoring and recommendation logic, and web application deployment. Each stage is designed to build upon the previous step, ensuring consistency, accuracy, and user relevance throughout the pipeline. The overall objective is to transform raw and incomplete childcare pricing data into meaningful, personalized insights that can support families in evaluating cost-effective childcare options across U.S. counties. Visualization techniques are integrated across stages to reveal data patterns, model interpretability, and enhance user experience. 

Starting with cleaning and exploring the childcare pricing data, the project then develops predictive models to estimate missing prices. These predictions feed into an affordability scoring system tailored to individual household profiles, which ultimately informs the recommendation engine.

The final phase focuses on delivering these insights via an interactive web application that allows users to explore affordability rankings and visualize geographic trends in childcare costs. It is important to note that the project scope remains flexible and may be adjusted depending on available time, resources, or emerging challenges to ensure feasible and meaningful outcomes.

## Data Preprocessing.


The preprocessing stage ensures that the data used for modeling and recommendation is clean, consistent, and analytically meaningful. The primary dataset, the National Database of Childcare Prices (NDCP), includes both observed and imputed childcare prices along with county-level demographic indicators. Key steps in the preprocessing phase include:

* **Year selection:** A single year (e.g., 2018) will be selected to avoid inconsistencies caused by cross-year data variability and shifting economic conditions.
* **Data cleaning:** Counties with incomplete socioeconomic indicators or extreme pricing outliers will be flagged and reviewed. Duplicate or redundant records will be removed.
* **Imputation management:** The NDCP includes flags identifying whether price data was directly collected or modeled. These flags will be used to filter out highly imputed values from the training set to ensure that predictions are built on reliable ground truth.
* **Feature transformation:** Continuous variables such as median household income and rent may be normalized or log-transformed to reduce skewness and improve model performance.
* **Feature engineering:** New variables will be created to support model interpretability and downstream ranking. These include affordability ratios (weekly price relative to household income) and categorical encodings for provider types and child age groups.
* **Exploratory data visualization:** Initial visualizations, such as histograms, boxplots, and scatter plots, will be generated to explore variable distributions, identify outliers, and reveal correlations between socioeconomic factors and childcare costs. Geographic heatmaps will visualize price variation across counties to provide spatial insights.

  
The processed dataset will serve as the foundation for model training and affordability analysis.

## Predictive Modelling.


The predictive modeling component is designed to estimate weekly childcare prices for counties where direct data is unavailable or incomplete. These price predictions are essential to providing full geographic coverage within the affordability recommender.

* **Target variable:** The model will predict the median weekly childcare price for a selected provider type (e.g., center-based care) and age group (e.g., preschoolers).
  
* **Input features:** Predictor variables include county-level indicators such as median household income, poverty rate, education levels, labor force participation, median gross rent, and urbanicity classification.
  
* **Model types:**
    * Linear regression and Ridge regression will be used as baseline models for initial testing and interpretability.
    * Tree-based ensemble models such as Random Forest and XGBoost will be evaluated for improved accuracy and robustness.
      
* **Evaluation metrics:** Model performance will be assessed using Root Mean Squared Error (RMSE), Mean Absolute Error (MAE), and R-squared (R²) on cross-validated test sets.
The best-performing model will be selected for use in generating price predictions for counties with missing or unreliable pricing data.

## Affordability and Scoring Recommendation

The recommender system transforms predicted and observed childcare prices into actionable insights by calculating affordability scores based on user input. These scores are then used to rank counties and suggest more affordable locations.

* **User-defined parameters:**
    *	Household income
    *	Child age group (e.g., toddler, preschooler)
    *	Provider type (center-based or home-based)
* **Affordability score calculation:**
    *	Each county’s affordability score is calculated as the ratio of median weekly childcare cost to household income.
    *	Counties with lower scores are considered more affordable relative to the user's financial profile.
* **Ranking logic:**
    *	Counties are ranked in ascending order of affordability score.
    *	Filters may be applied to restrict results to specific states or regions.
    *	Optional display of percentile-based affordability tiers or affordability thresholds may be included for interpretability.

This component translates raw pricing and demographic data into personalized recommendations that align with users' budgets and care needs.

## Handling Missing Values with Imputation Strategies

* Defines a helper function to display columns with over 20% missing data to target for imputation or removal.
* Imputes missing numeric values by filling with column medians, a robust approach to handle missingness without skewing distributions significantly.

```python
# Checking count of missing values in each column
missing_summary = SmartCare_df.isnull().sum().sort_values(ascending=False)
missing_summary[missing_summary > 0]
```

```python
# Creating a function to find the % of missing values in dataframe
# It takes data input, counts the sum of null values in a column, computes its percent value, filters all columns with above 20%, and return the list of columns
def get_NArate(input_data):
    columns = (input_data.isnull().sum() * 100/len(input_data)).sort_values(ascending=False).round(2)
    columns = columns[abs(columns) >= 20]
    print(columns)
```

```python
get_NArate(SmartCare_df)
```

```python
# Getting the list of columns with numeric data
numcols = SmartCare_df.select_dtypes('number').columns.tolist()
```

```python
# Using a for loop to find missing values in numeric columns and replace the values with that column's median value
for column in SmartCare_df[numcols]:
    median = SmartCare_df[numcols].median()
    SmartCare_df[numcols] = SmartCare_df[numcols].fillna(median)
```

Missing data is quantified, with a focus on columns exhibiting more than 20% missingness. Median imputation is applied to all numeric variables containing missing values, replacing missing entries with the median of the respective column. This approach is less sensitive to outliers than mean imputation, preserving the dataset's central tendency.

# Feature Engineering and Transformation

## Create Derived Variables

To enhance model effectiveness and exploratory analysis, new features are created:

* Age Group Classification: Based on which price columns have data, a categorical age_group is assigned (infant, toddler, preschool, or unspecified). This helps in segmenting childcare types by age for targeted analysis.
* Care Type Identification: A categorical care_type is created to indicate whether a record contains center-based care data, family-based care data, or both (mixed). This differentiates types of childcare providers.
* Affordability Scores: New columns calculate affordability by dividing weekly childcare prices by median household income for each age group and care type. This normalized metric helps compare childcare cost relative to local income levels, providing a fair affordability measure.
* Interaction Term for Poverty Burden: Multiplying poverty rate by unemployment rate creates a composite socioeconomic stress indicator, which may improve model explanatory power.

```python
# Creating a function to derive Age Group from Childcare Type Columns 
def get_age_group(row):
    if pd.notnull(row['price_center_infant']) or pd.notnull(row['price_family_infant']):
        return 'infant'
    elif pd.notnull(row['price_center_toddler']) or pd.notnull(row['price_family_toddler']):
        return 'toddler'
    elif pd.notnull(row['price_center_preschool']) or pd.notnull(row['price_family_preschool']):
        return 'preschool'
    else:
        return 'unspecified'
```

```python
# Calling the get_age_group to create a new datafield
SmartCare_df['age_group'] = SmartCare_df.apply(get_age_group, axis=1)
```

```python
# Creating a function to derive care type from the center-based and family-based childcare pricing fields
def get_care_type(row):
    has_center = any(pd.notnull(row[col]) for col in [
        'price_center_infant', 'price_center_toddler', 'price_center_preschool'
    ])
    has_family = any(pd.notnull(row[col]) for col in [
        'price_family_infant', 'price_family_toddler', 'price_family_preschool'
    ])

    if has_center and has_family:
        return 'mixed'
    elif has_center:
        return 'center'
    elif has_family:
        return 'family'
    else:
        return 'unspecified'
```

```python
# Calling the get_care_type to create a new datafield
SmartCare_df['care_type'] = SmartCare_df.apply(get_care_type, axis=1)
```

```python
# Calculating the Affordability score as = Weekly price / Household income
# Center-based affordability scores
SmartCare_df['afford_center_infant'] = SmartCare_df['price_center_infant'] / SmartCare_df['median_household_income']
SmartCare_df['afford_center_toddler'] = SmartCare_df['price_center_toddler'] / SmartCare_df['median_household_income']
SmartCare_df['afford_center_preschool'] = SmartCare_df['price_center_preschool'] / SmartCare_df['median_household_income']

# Family-based affordability scores
SmartCare_df['afford_family_infant'] = SmartCare_df['price_family_infant'] / SmartCare_df['median_household_income']
SmartCare_df['afford_family_toddler'] = SmartCare_df['price_family_toddler'] / SmartCare_df['median_household_income']
SmartCare_df['afford_family_preschool'] = SmartCare_df['price_family_preschool'] / SmartCare_df['median_household_income']
```

```python
# calculating the Poverty burden as: Poverty rate x unemployment rate (optional interaction term)
SmartCare_df['poverty_unemployment_interaction'] = SmartCare_df['poverty_rate'] * SmartCare_df['unemployment_rate']
```

# Predictive Modeling and Evaluation

Predictive modeling constitutes the core analytical phase, where the project develops and compares machine learning models to estimate childcare costs. This section explains the preprocessing steps such as feature scaling and dimensionality reduction, followed by the implementation of baseline and advanced regression techniques. It describes model training, testing, and rigorous evaluation using standard metrics to select the best-performing models.

## Data Standardization and PCA Transformation

Before modeling, features (excluding identifiers and target variables) are standardized to have zero mean and unit variance. PCA is then applied to reduce dimensionality while retaining 90% of data variance, which helps reduce noise and computational complexity.

```python
# Selecting features for modeling by dropping identifiers and target variables (exclude ID/state columns, target, etc.)
features = SmartCare_df.drop(columns=['state', 'county', 'fips', 'year','price_center_infant','price_family_infant'])
```

```python
# Standardizing features to zero-mean and unit-variance
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
```

```python
# Applying PCA to reduce dimensionality while preserving 90% of variance
pca_obj = PCA(n_components=0.90, whiten=True,random_state=42)
X_pca = pca_obj.fit_transform(X_scaled)
```

```python
# Defining target variables for both care types
y1 = SmartCare_df['price_center_infant']  # Center-based
y2 = SmartCare_df['price_family_infant']  # Family-based
```

```python
# Applying Train-test split for each target
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_pca, y1, test_size=0.2, random_state=42)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_pca, y2, test_size=0.2, random_state=42)
```

## Baseline Linear Regression

Linear regression models are trained separately to predict center-based and family-based infant care prices. The data is split into training and testing sets to evaluate model generalization. Standard regression metrics (R², MAE, MSE, RMSE) provide baseline performance benchmarks.

```python
# Training a model
baseline_model1 = LinearRegression()
baseline_model2 = LinearRegression()
baseline_model1.fit(X_train1, y_train1)
baseline_model2.fit(X_train2, y_train2)
y1_pred_lr = baseline_model1.predict(X_test1)
y2_pred_lr = baseline_model2.predict(X_test2)
```

```python
# Creating a universal Metrics function
def Calc_Metrics(true, pred):
    R2 = r2_score(true, pred).round(3)
    MAE = mean_absolute_error(true, pred).round(4)
    MSE = mean_squared_error(true, pred).round(4)
    RMSE = np.sqrt(mean_squared_error(true, pred)).round(4)
    return R2, MAE, MSE, RMSE
```

```python
# Evaluating the baseline model 
r2_lr1, mae_lr1, mse_lr1, rmse_lr1 = Calc_Metrics(y_test1, y1_pred_lr)
print(f"Linear Regression R²: {r2_lr1}")
print(f"MAE: {mae_lr1}, MSE: {mse_lr1}, RMSE: {rmse_lr1}")
```

```python
# Calculating baseline model metrics
r2_lr2, mae_lr2, mse_lr2, rmse_lr2 = Calc_Metrics(y_test2, y2_pred_lr)
print(f"Linear Regression R²: {r2_lr2}")
print(f"MAE: {mae_lr2}, MSE: {mse_lr2}, RMSE: {rmse_lr2}")
```

## Advanced Regression Pipelines

A set of more advanced regression models is evaluated, including Ridge, Lasso, Elastic Net, K-Nearest Neighbors, Decision Tree, Random Forest, and Gradient Boosting. Each model is wrapped in a pipeline for consistency, trained, and tested, enabling performance comparisons.

```python
# Defining model dictionary
regressors = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(random_state=42),
    'Lasso Regression': Lasso(random_state=42),
    'Elastic Net': ElasticNet(random_state=42),
    'K-Nearest Neighbor': KNeighborsRegressor(),
    'Decision Tree Regressor': DecisionTreeRegressor(random_state=42),
    'Random Forest Regressor': RandomForestRegressor(random_state=42),
    'Gradient Boosting Regressor': GradientBoostingRegressor(random_state=42)
}
```

```python
reg_result_y1 = {}
reg_result_y2 = {}

def build_pipeline(model):
    return Pipeline([('regressor', model)])

def train_and_eval(X_train, X_test, y_train, y_test, model):
    pipeline = build_pipeline(model)
    trained = pipeline.fit(X_train, y_train)
    pred = trained.predict(X_test)
    r2, mae, mse, rmse = Calc_Metrics(y_test, pred)
    acc = trained.score(X_test, y_test).round(2)
    return r2, mae, mse, rmse, acc

# Evaluating all models for both y1 and y2
for name, model in regressors.items():
    print(f"\nTraining {name} on center-based prices...")
    r2, mae, mse, rmse, acc = train_and_eval(X_train1, X_test1, y_train1, y_test1, model)
    reg_result_y1[name] = {'R2': r2, 'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'Accuracy': acc}

    print(f"Training {name} on family-based prices...")
    r2, mae, mse, rmse, acc = train_and_eval(X_train2, X_test2, y_train2, y_test2, model)
    reg_result_y2[name] = {'R2': r2, 'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'Accuracy': acc}
```

```python
# Creating the main Functioncalled smartcare_recommender. 
# It recommends the most affordable counties for childcare based on user income, care type, age group, and optional state filter.
def smartcare_recommender(user_income, user_state=None, top_n=5, show_plot=True):
    try:
        # Validate that income input is a positive value
        if user_income <= 0:
            raise ValueError("Income must be greater than 0.")

        # Define the two types of care and three age categories
        care_types = ['center', 'family']
        age_groups = ['infant', 'toddler', 'preschool']

        # Initialize a dictionary to store top results for each care type
        records = {'Center': [], 'Family': []}

        # Iterate through each combination of care type and age group
        for care in care_types:
            for age in age_groups:
                col_name = f'price_{care}_{age}'  # Construct column name dynamically

                # Skip if this price column is not in the dataset
                if col_name not in SmartCare_df.columns:
                    continue

                # Extract relevant columns and remove rows with missing prices
                temp_df = SmartCare_df[['state', 'county', col_name]].copy()
                temp_df = temp_df.dropna()

                # Calculate affordability as price divided by user income
                temp_df['Affordability'] = (temp_df[col_name] / user_income).round(4)

                # Rename columns for display clarity
                temp_df = temp_df.rename(columns={
                    'state': 'State',
                    'county': 'County',
                    col_name: 'Price'
                })

                # Add labels for care type and age group
                temp_df['Age Group'] = age.title()
                temp_df['Care Type'] = care.title()

                # If user provides a state, filter results to that state
                if user_state:
                    input_state = user_state.strip().upper()
                    full_state = state_abbrev_to_full.get(input_state, None)

                    # If no abbreviation match, attempt direct name match
                    if full_state is None:
                        full_state = user_state.strip().title()

                    # If the state is invalid or not found, skip this iteration
                    if full_state not in SmartCare_df['state'].unique():
                        continue

                    # Filter the dataframe to include only rows for the selected state
                    temp_df = temp_df[temp_df['State'] == full_state]

                # Append the top N affordable counties to the appropriate care list
                records[care.title()].append(temp_df.sort_values(by='Affordability').head(top_n))

        # Concatenate individual DataFrames for both care types
        center_df = pd.concat(records['Center'], ignore_index=True)
        family_df = pd.concat(records['Family'], ignore_index=True)

        # If no results are found for either care type, exit with a warning
        if center_df.empty and family_df.empty:
            print("⚠️ No results found for the selected input.")
            return

        # Determine display label for state
        display_state = full_state if user_state else "All States"

        # Display summary information
        print(f"\033[1m\nState: {display_state} |  Income: ${user_income:,} / week\033[0m\n")

        # Helper function to format and print results
        def display_table(df, title):
            # Extract and format columns
            df_clean = df[['County', 'Age Group', 'Price', 'Affordability']].copy()
            df_clean['Price'] = df_clean['Price'].apply(lambda x: f"${x:.2f}")
            df_clean['Affordability'] = df_clean['Affordability'].apply(lambda x: f"{x:.3f}")

            # Display formatted table
            print(f"\033[1m{title} Care:\033[0m\n")
            print(df_clean.to_string(index=False))
            print()

        # Display center-based recommendations
        if not center_df.empty:
            display_table(center_df, "Center-Based")

        # Display family-based recommendations
        if not family_df.empty:
            display_table(family_df, "Family-Based")

    # Catch and display any errors during execution
    except Exception as e:
        print(f"⚠️ Error: {e}")
```

# Discussion and Summary of Key Findings

The SmartCare Locator project successfully integrates machine learning with socio-economic data to generate actionable childcare cost recommendations at the county level. Through robust preprocessing and feature engineering, we distilled over 300 variables into a focused set of predictors that drive affordability estimation and recommendations.

The predictive models achieved high accuracy, with Random Forest and Gradient Boosting regressors outperforming linear models, especially for family-based care. The PCA-based pipeline reduced dimensionality while retaining interpretability. Exploratory data analysis uncovered substantial regional disparities in pricing and affordability, and racial demographic patterns also provided insights into socio-economic disparities.

Assumptions made during the project include:

* Median weekly prices adequately represent typical costs for families.
* Demographic and economic features used as predictors are stable and reflect actual cost drivers.
* Affordability can be reasonably modeled as a simple ratio of weekly price to household income.

These assumptions allowed for a generalizable affordability scoring logic and model training process that serves the intended purpose of the tool.

#  Limitations and Challenges

This project, while robust, encountered several limitations. First, the NDCP dataset, though comprehensive, contained many missing values and inconsistencies, particularly for certain counties or age/care combinations. Imputation strategies were applied, but residual data sparsity may still affect prediction fidelity in underrepresented regions.

Key challenges included:

* Feature redundancy and high correlation, which required dimensionality reduction to ensure model stability.
* Generalization to unseen counties, especially rural ones with unique characteristics, may be imperfect.
* Simplified affordability logic, which does not yet account for multi-child households, subsidies, or availability of care providers.

Technical challenges were also encountered in balancing model performance with interpretability. Tree-based models provided better predictions but are less transparent than linear alternatives.

# Future Enhancements and Recommendations

Several directions exist to enhance the current solution:

* User Interface Deployment: Implement the affordability recommender in a production-ready Streamlit application, allowing real-time user interaction.
* Filter and Personalization Enhancements: Add filters for multiple children, household size, and provider quality ratings if available.
* Model Expansion: Extend modeling to additional age groups and years, or forecast future prices based on macroeconomic trends.
* Affordability Mapping: Integrate with geographic visualization tools (e.g., Folium, Plotly) to allow users to explore cost landscapes visually.

It is recommended that any policy-related use of this tool be coupled with validation against newer local market rate surveys. For families, this tool can already provide a directional insight into childcare affordability and aid in location-specific decisions.

# Ethical Considerations

Ethics played a critical role in shaping the project’s methodology. While the goal was to inform users and policymakers, care was taken not to imply that affordability alone should dictate childcare decisions, which may also involve quality, accessibility, and cultural preferences.

Other ethical considerations include:

* Privacy: No personally identifiable information (PII) was used.
* Bias in Predictions: The underlying data may reflect structural inequalities. Predictive models must be interpreted with this in mind, as affordability in one region does not imply equity or adequacy of care.
* Misuse Prevention: Any future public-facing deployment must clearly communicate that affordability scores are indicative, not prescriptive.

Efforts should continue to include more community- and provider-level information to ensure socially responsible guidance.
